Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
movies = pd.read_csv("/content/MoviesOnStreamingPlatforms_updated.csv")

# Remove the "%" symbol and convert "Rotten Tomatoes" to float
movies['Rotten Tomatoes'] = movies["Rotten Tomatoes"].str.replace("%", "").astype(float)

# Drop non-numeric columns
movies_numeric = movies.select_dtypes(include=['float64', 'int64'])

# Calculate the correlations
correlations = movies_numeric.corr(method='pearson')

# Generate the heatmap
sns.heatmap(correlations, cmap="coolwarm")
plt.show()
```

```python
sns.heatmap(correlations, cmap="coolwarm", annot=True)
plt.show()
```

Pie Chart

```python
group_size = [sum(df["sepal length (cm)"]), sum(df["sepal width (cm)"]),
              sum(df["petal length (cm)"]), sum(df["petal width (cm)"])]
group_labels = ["sepal length (cm\n" + str(sum(df["sepal length (cm)"])) + "\n",
                "sepal width (cm)\n" + str(sum(df["sepal width (cm)"])) + "\n",
                "petal length (cm)\n" + str(sum(df["petal length (cm)"])) + "\n",
                "petal width (cm)\n" + str(sum(df["petal width (cm)"]))]
custom_colors = ["skyblue", "yellowgreen", "tomato", "red"]
plt.figure(figsize=(5, 5))
plt.pie(group_size, labels=group_labels, colors=custom_colors)
plt.rc('font', size=12)
plt.title("Total sepal length (cm),sepal width (cm),petal length (cm)	and petal width (cm) ", fontsize=20)
plt.show()
```

Donut plot

```python
group_size = [sum(df["sepal length (cm)"]), sum(df["sepal width (cm)"]),
              sum(df["petal length (cm)"]), sum(df["petal width (cm)"])]
group_labels = ["sepal length (cm\n" + str(sum(df["sepal length (cm)"])) + "\n",
                "sepal width (cm)\n" + str(sum(df["sepal width (cm)"])) + "\n",
                "petal length (cm)\n" + str(sum(df["petal length (cm)"])) + "\n",
                "petal width (cm)\n" + str(sum(df["petal width (cm)"]))]
custom_colors = ["skyblue", "yellowgreen", "tomato", "red"]
plt.figure(figsize=(5, 5))
plt.pie(group_size, labels=group_labels, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Total sepal length (cm),sepal width (cm),petal length (cm)	and petal width (cm) ", fontsize=20)
plt.show()
```

Visualizing a Treemap using Python


```python
import plotly.graph_objects as go
import pandas as pd
from sklearn import datasets

# Load Iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# Create labels and parents for the treemap
labels = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]
parents = ["Setosa", "Versicolor", "Virginica"] * len(labels)

# Plot
fig = go.Figure(go.Treemap(
    labels=labels,
    parents=parents
))

fig.show()
```

Plotting Annotations using Python

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets

# Load Iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# Create labels and parents for the treemap
x = df['species']
y = df['sepal length (cm)']

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y)
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Length Distribution Across Species')
plt.show()
```

Word Cloud

```python
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_excel("/content/Instacomments.xlsx")
print(data.head())
```

```python
text = " ".join(i for i in data.text)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

Animated Scatter Plot using Python

```python
import pandas as pd
import plotly.express as px
from sklearn import datasets

# Load Iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# Add artificial time dimension
df['time'] = range(len(df))

# Create animated scatter plot
fig = px.scatter(df,
                 x="sepal length (cm)",
                 y="sepal width (cm)",
#                 z = "petal length (cm)",
 #                ="petal width (cm)",
                 animation_frame="time",
                 animation_group="species",
                 color="species",
                 hover_name="species",
                 title="Animated Scatter Plot of Iris Dataset")
fig.show()
```

```python
from sklearn import datasets

iris = datasets.load_iris()
df_iris = pd.DataFrame(iris['data'],
                       columns=iris['feature_names'])
df_iris['target'] = iris['target']
df_iris.head()
```

```python
import graphviz
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Train a decision tree classifier
clf = DecisionTreeClassifier()
clf.fit(X, y)

# Export the decision tree as a Graphviz object
dot_data = export_graphviz(clf, out_file=None,
                     feature_names=iris.feature_names,
                     class_names=iris.target_names,
                     filled=True, rounded=True,
                     special_characters=True)

# Render the Graphviz object
graph = graphviz.Source(dot_data)
graph
```
