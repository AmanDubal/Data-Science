```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

```python
X= load_wine().data
y = load_wine().target
```

```python

# Apply DBSCAN clustering
dbscan_clustering = DBSCAN(eps=0.8, min_samples=5)
dbscan_labels = dbscan_clustering.fit_predict(X)
# Plot the original data points with colors representing the ground truth clusters
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='Blues', marker='o', edgecolors='k', label='Ground Truth')
# Plot the data points with colors representing the predicted clusters by DBSCAN
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='hot_r', marker='s', edgecolors='k', alpha=0.5, label='DBSCAN Predicte')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```


#  DBSCAN Clustering on Wine Dataset

# Introduction:
This experiment demonstrates the application of DBSCAN clustering on the wine dataset. The objective is to identify distinct clusters within the data based on their inherent characteristics.

# Methodology:
1. Data Import:
   - The wine dataset is imported using the `load_wine()` function from sklearn.datasets.
   - The data features are stored in the `X` variable, and the ground truth labels are stored in the `y` variable.

2. DBSCAN Clustering:
   - DBSCAN clustering is performed on the data using the `DBSCAN` class from sklearn.cluster.
   - The parameters `eps=0.8` and `min_samples=5` are chosen for the clustering.
   - The predicted cluster labels are stored in the `dbscan_labels` variable.

3. Visualization:
   - The original data points are plotted with colors representing the ground truth clusters.
   - The data points are also plotted with colors representing the predicted clusters by DBSCAN.
   - The plot is titled 'DBSCAN Clustering' and includes labels for the axes and a legend.

# Results:
The plot displays the original data points and the predicted clusters by DBSCAN. By comparing the two plots, we can observe the effectiveness of DBSCAN in identifying distinct clusters within the data.

# Conclusion:
This experiment demonstrates the successful application of DBSCAN clustering on the wine dataset. The results indicate that DBSCAN can effectively identify clusters in the data, even when the data is noisy or has complex shapes.
