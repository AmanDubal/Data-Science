```python
import plotly.graph_objects as go
import plotly.io as pio

# Set the default template for the plot
pio.templates.default = "plotly_white"

# Define the funnel stages
funnel_stages = ['From Home', 'From Hashtags', 'From Explore', 'From Other', 'Saves']

# Calculate the number of users and conversions for each stage
num_users = []
num_conversions = []

for stage in funnel_stages:
    stage_users = data[stage]  # Access the column using the stage name
    num_users.append(len(stage_users))
    num_conversions.append(stage_users.sum())

# Create a funnel chart
fig = go.Figure(go.Funnel(
    y=funnel_stages,
    x=num_users,
    textposition='inside',
    textinfo='value',
    name='Users'
))

fig.add_trace(go.Funnel(
    y=funnel_stages,
    x=num_conversions,
    textposition='inside',
    textinfo='value',
    name='Conversions'
))

fig.update_layout(
    title='Funnel Analysis',
    funnelmode='stack'
)

fig.show()
```
